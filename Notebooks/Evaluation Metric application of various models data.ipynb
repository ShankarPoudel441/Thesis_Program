```python
X=np.array(df[["p0","p1","p2","p3","p4","p5","p6"]])
y=np.array(df["cluster"])
```

```python
from sklearn.metrics import silhouette_score
#Higher Better
#[-1,1]
#Very near to zero, overlapping clusters
sihl = silhouette_score(X, y, metric="euclidean")
sihl
```

```python
from sklearn.metrics import calinski_harabasz_score
#Higher Better
calin = calinski_harabasz_score(X, y)
calin
```

```python
from sklearn.metrics import davies_bouldin_score
#Lower better
DB = davies_bouldin_score(X, y)
DB
```

```python
def intrinsic_metrics(X,clusters):
    print("")
    sihl = silhouette_score(X, clusters, metric="euclidean")
    calin = calinski_harabasz_score(X, clusters)
    DB = davies_bouldin_score(X, clusters)
    print(f"sihl={sihl}\ncalin={calin}\nDB={DB}\n")
    return sihl,calin,DB

def intrinsic_metrics_betn_class(X,clusters,class1,class2):
    print("")
    sihl = silhouette_score(X, clusters, metric="euclidean")
    calin = calinski_harabasz_score(X, clusters)
    DB = davies_bouldin_score(X, clusters)
    print(f"sihl={sihl}\ncalin={calin}\nDB={DB}\n")
    return sihl,calin,DB
```
